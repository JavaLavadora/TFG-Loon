```python
import pandas as pd
import numpy as np
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#['N211LB', 'N235LB', 'N238LB', 'N252LB', 'N271LB', 'N329LB']

fr = pd.DataFrame()

loons_to_study = ['N211LB', 'N235LB', 'N238LB', 'N252LB', 'N271LB', 'N329LB']

for loon in loons_to_study:
    
    aux = pd.read_csv('../../../../og_data/ST-DBSCAN/ST-' + loon + '.csv', parse_dates=['ts'], \
                       low_memory=False)
   
    fr = fr.append(aux)

fr['trip_id'] = fr['registration_id']

bl = fr
bl = bl.sort_values('ts')
bl['visit_id'] = bl['registration_id'].str.cat(bl['cluster'].astype(str), sep='/')\
                                        .str.cat(bl['episode'].astype(str), sep='/')
```

```python
# As expected the absolute values do not show a correlation with the stationarity due to the way the loons move
# they control the distance they move by choosing the time they will stay on a wind stream therefore even when they 
# are providing service they might take a stronger wind that suits their intended trajectory for a short period 
# of time
```

```python
# Do loons use roads when they want to move?

# We study only moving traces since that's when the loon will look for roads
road_loons = bl.loc[bl.stationary == False].copy()

road_loons['climate'] = pd.cut(road_loons['lat'], bins=range(-90, 90, 10), include_lowest=True)

road_loons['meridian'] = pd.cut(road_loons['lon'], bins=range(-180, 180, 10), include_lowest=True)

# Let's plot the tendencies of the loons to see where they like to be
```

```python
# Plot latitude
plt.figure(figsize=(20,6))

road_loons_alt_point_count = road_loons.groupby('climate').count().reset_index()

ax = sns.barplot(x=road_loons_alt_point_count['climate'], y="speed", \
                data=road_loons_alt_point_count)

dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=17)
```

```python
# Plot longitude
plt.figure(figsize=(20,6))

road_loons_lon_point_count = road_loons.groupby('meridian').count().reset_index()

ax = sns.barplot(x=road_loons_lon_point_count['meridian'], y="speed", \
                data=road_loons_lon_point_count)

dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=17)
```

```python
# The loons tend to move near the ecuator (-25, 25) and on southamerican meridians 
#(this is due to most of the loons operating in southamerica). 
```

```python
# How does the velocity vary?

#Plot latitude
plt.figure(figsize=(20,6))

road_loons_alt_mean_velocity = road_loons.groupby('climate').mean().reset_index()

ax = sns.barplot(x=road_loons_alt_mean_velocity['climate'], y="speed", \
                data=road_loons_alt_mean_velocity)
ax.set(ylim=(0, 50))
dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=17)
```

```python
#Plot longitude

plt.figure(figsize=(20,6))

road_loons_lon_mean_velocity = road_loons.groupby('meridian').mean().reset_index()

ax = sns.barplot(x=road_loons_lon_mean_velocity['meridian'], y="speed", \
                data=road_loons_lon_mean_velocity)
ax.set(ylim=(0, 50))
dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=17)
```

```python
# ---------------------------------------Altitude study---------------------------------------
```

```python
# Hypothesis is that altitudes are not related to stationarity for two reasons:
# 1. The loon changes altitude to look for suitable winds which are stochastic
# 2. The loon lives on the stratosphere where there are no climate phenomenon
# But still there might be some exceptions (the loon might go down to provide better coverage 
# or to receive instructions)

loons_by_mean_altitude = bl.groupby(['registration_id', 'stationary'])['alt'].mean().reset_index().copy()


ax = sns.barplot(x=loons_by_mean_altitude['registration_id'], y="alt", hue='stationary',\
                data=loons_by_mean_altitude)

dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=7)
```

```python
# As expected on average there is no significant altitude difference between a moving and a stay loon
# Let's check absolute values


loons_by_mean_altitude_max = bl.groupby(['registration_id', 'stationary'])['alt'].max().reset_index().copy()


ax = sns.barplot(x=loons_by_mean_altitude_max['registration_id'], y="alt", hue='stationary',\
                data=loons_by_mean_altitude_max)

dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=7)
```

```python
loons_by_mean_altitude_min = bl.groupby(['registration_id', 'stationary'])['alt'].min().reset_index().copy()


ax = sns.barplot(x=loons_by_mean_altitude_min['registration_id'], y="alt", hue='stationary',\
                data=loons_by_mean_altitude_min)

dummy = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
dummy = ax.tick_params(labelsize=7)
```

```python
# Service times
```

```python
# First get a bl with the stay points traces to check service time
bl_sp = bl.loc[bl.stationary == True]

stationary_by_episode_start = bl_sp.groupby(['cluster','episode'])['ts'].first().reset_index()

stationary_by_episode_end = bl_sp.groupby(['cluster','episode'])['ts'].last().reset_index()

stationary_by_episode_duration = stationary_by_episode_start[['cluster', 'episode']]
stationary_by_episode_duration['start'] = stationary_by_episode_start['ts']
stationary_by_episode_duration['end'] = stationary_by_episode_end['ts']
stationary_by_episode_duration['duration'] = stationary_by_episode_duration['end'] \
                                            -stationary_by_episode_duration['start']

total_service_time = stationary_by_episode_duration['duration'].sum()

# For the total time we need a reference so we use the first record and compute the time delt
flight_times = bl['ts'] - bl['ts'].shift()
flight_times = flight_times.fillna(pd.Timedelta(seconds=0))

total_flight_time = flight_times.sum()

# Let's compute the ratio
ratio = total_service_time / total_flight_time
ratio
```
