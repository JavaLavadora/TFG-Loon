```python
import pandas as pd
import numpy as np

from keplergl import KeplerGl
import geopandas as gpd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#['N139LB', 'N166LB', 'N191LB', 'N211LB', 'N221LB', 'N225LB', 'N226LB', 'N228LB', 'N234LB',
#                   'N235LB', 'N238LB', 'N252LB', 'N253LB', 'N271LB', 'N329LB', 'N789LB']


raw_df = pd.DataFrame()

flights_to_add =  ['N139LB', 'N166LB', 'N191LB', 'N211LB', 'N221LB', 'N225LB', 'N226LB', 'N228LB', 'N234LB',
                  'N235LB', 'N238LB', 'N252LB', 'N253LB', 'N271LB', 'N329LB', 'N789LB']

for loon in flights_to_add:

    aux = pd.read_csv('../../../../og_data/' + loon + '.csv', parse_dates=['ts'],\
                          low_memory=False)
    
    raw_df = raw_df.append(aux)

```

```python
bl = raw_df
bl
```

```python


plt.figure(figsize=(20,6))
plt.xticks(rotation=90)

#ax = sns.distplot(bl.loc[bl.registration_id == loon].speed)

ax = sns.boxenplot(x="registration_id", y="speed", scale="linear", data=bl)

#ax.set(ylim=(0, 50))
dummy = ax.set_xlabel('Registration Id',fontsize=17)
dummy = ax.set_ylabel('Velocity (Km/h)',fontsize=17)
dummy = ax.set_title('Velocity value distribution of the captured loons')

dummy = ax.tick_params(labelsize=17)
```

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=90)

aux = bl.loc[(bl.registration_id == 'N221LB')].sort_values('ts').tail(1500)
#aux['ts_str'] = aux['ts'].astype(str)
ax = sns.lineplot(x="ts", y="alt", hue="registration_id", data=aux)

dummy = ax.set_xlabel('Timestamp of year 2020 (Month-Day Hour)',fontsize=17)
dummy = ax.set_ylabel('Altitude (feet)',fontsize=17)
dummy = ax.set_title('Velocity value distribution of the captured loons')

dummy = ax.tick_params(labelsize=17)
```
