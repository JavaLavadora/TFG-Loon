```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_df =  pd.read_csv("../../../../og_data/extract_clean_with_country.csv", index_col='Unnamed: 0', parse_dates=['ts'])
```

```python
# We must therefore understand what defines a flight from another
bl = raw_df
```

```python
flights_per_loon = pd.DataFrame()
flights_per_loon = bl[['aircraft_id', 'flight_id']].groupby('aircraft_id').nunique()

flights_per_loon['flight_count'] = flights_per_loon['flight_id']
flights_per_loon = flights_per_loon.drop(columns=['aircraft_id', 'flight_id'])
flights_per_loon = flights_per_loon.reset_index()
flights_per_loon = flights_per_loon.sort_values(by='flight_count', ascending=False)
```

```python
flights_per_loon.head(10)
```

```python
def find_flight_changes(id):
    #Get the loon record info
    loon_record = bl.loc[bl.aircraft_id == id]
    #loon_record['bl_index'] = loon_record.index
    #Get index for first and last record of each flight
    #flight_start = loon_record[['flight_id', 'bl_index']].groupby('flight_id').max().reset_index()
    #flight_end = loon_record[['flight_id', 'bl_index']].groupby('flight_id').min().reset_index()
    flight_start = loon_record.groupby('flight_id').first().reset_index()
    flight_end = loon_record.groupby('flight_id').last().reset_index()
    
    #Combine into ordered df
    flight_start_end = pd.concat([flight_start, flight_end]).sort_values(by=['flight_id'])
    
    return flight_start_end
```

```python
# Let's plot the data for the 3 loons with most flights
# First the velocities
traveller_loon_1 = flight_changes.loc[flight_changes.aircraft_id == flights_per_loon['aircraft_id'].iloc[0]] \
                                 .sort_values(by='ts').reset_index(drop=True)

traveller_loon_1.loc[(traveller_loon_1.index % 2) == 0, 'flight_stage'] = 'START'
traveller_loon_1.loc[(traveller_loon_1.index % 2) == 1, 'flight_stage'] = 'FINISH'

plt.figure(figsize=(20,6))

ax = sns.barplot(x=traveller_loon_1['ts'], y="speed", hue='flight_stage', \
            hue_order=['FINISH', 'START'] , data=traveller_loon_1)

b = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```
