```python
import pandas as pd
import numpy as np
```

```python
# Load the loon data to sample
raw_df = pd.read_csv('../../../../og_data/extract.csv', names=['craw_ts', 'id', 'int_id', 'lat', 'lon', 'azi', 'alt', 'speed', 'ts'],
                         low_memory=False)
```

```python
# Base line is the dataframe from which we can start studying the cleaned data
bl = raw_df
```

```python
# Leave faulty data out of the base line
# dfByAlt = raw_df.groupby(by='alt').count().reset_index()
# dfByAlt.to_csv('data/loonsByAltitude.csv')
bl = bl.query(' alt!="HBAL080" & alt!="A38D04" ')
bl = bl.query(' ts!="HBAL399" ')

#Clarify column names (int_id -> aircraft_id; id -> flight_id)
bl = bl.rename({'int_id': 'aircraft_id', 'id': 'flight_id'}, axis='columns')

# This one worked on mac but not on linux
bl['ts'] = pd.to_datetime(bl['ts'], format='%Y%m%d%H%M%S', errors='coerce')


# Numpy uses objects for strings
# Convert alt column to int
bl['alt'] = bl['alt'].astype(np.float)  
bl['trip_id'] = bl['flight_id'].str.cat(bl['aircraft_id'],sep="-")
```

```python
# Let's save the cleaned df
bl.to_csv(path_or_buf='../../../../og_data/extract_clean.csv')
```
