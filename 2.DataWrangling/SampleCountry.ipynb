```python
import pandas as pd
import numpy as np
import geopandas as gpd
import time
```

```python
#['N139LB', 'N166LB', 'N211LB', 'N221LB', 'N225LB', 'N226LB', 'N235LB', 'N238LB', 'N252LB',\
#                 'N271LB', 'N329LB', 'N789LB']

raw_df = pd.DataFrame()

flights_to_add = ['N139LB', 'N166LB', 'N191LB', 'N211LB', 'N221LB', 'N225LB', 'N226LB', 'N228LB', 'N234LB',
                   'N235LB', 'N238LB', 'N252LB', 'N253LB', 'N271LB', 'N329LB', 'N789LB']

for flight in flights_to_add:
    
    aux = pd.read_csv('../../../../og_data/ST-DBSCAN/ST-' + flight + '.csv', parse_dates=['ts'], low_memory=False)
    
    raw_df = raw_df.append(aux)
```

```python
# We load the file containing the coordinates of the country borders
world_gdf = gpd.read_file('../../../../og_data/countries.geojson')
```

```python
# Convert initial df to geodf and set the coordinates reference system to match the one in the borders file
bl_gdf = gpd.GeoDataFrame(raw_df, geometry=gpd.points_from_xy(raw_df.lon, raw_df.lat))
bl_gdf.crs = {'init': 'epsg:4326'}
```

```python
bl_gdf.head(10)
```

```python
# Perform a geopandas spatial join to check the country where each coordinate falls in
gdf_with_country = gpd.sjoin(bl_gdf, world_gdf, how='left')
```

```python
gdf_with_country
```

```python
# Improve readability of expanded dataset

bl_to_print = gdf_with_country
bl_to_print['country'] = gdf_with_country['ADMIN']
bl_to_print['country_code'] = gdf_with_country['ISO_A3']

bl_to_print = bl_to_print.drop(['geometry', 'index_right', 'ADMIN', 'ISO_A3', 'ISO_A2'], axis=1)

bl_to_print
```

```python
for flight in flights_to_add:

    bl_to_print.loc[bl_to_print.registration_id == flight].sort_values('ts')\
                    .to_csv(path_or_buf='../../../../og_data/Country/' + flight + '-country.csv', \
                    index=False, encoding='utf-8-sig')
```
